```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv("players_22.csv")
players
```

```python
features = ["overall", "potential", "wage_eur", "value_eur", "age"]
```

```python
# Data cleaning
players = players.dropna(subset=features)
```

```python
data = players[features].copy()
```

### Steps:

- Scale the values in the columns to ensure no one column dominates over the others
- Initialize random centroids
- Update centroids
- Iterate until centroids stop changing

```python
# Scale from 1-10 using min-max

data = ((data - data.min()) / (data.max() - data.min())) * 9 + 1
data.describe()
```

```python
data.head()
```

```python
def random_centroids(data, k):
    """Randomly generate *k* number of centroids based on randomly chosen values from the scale provided in the *data*"""
    centroids = []
    for i in range(k):
        centroid = data.apply(lambda x:float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

```python
centroids = random_centroids(data, 5)
centroids
```

```python
def get_labels(data, centroids):
    distances = centroids.apply(lambda x:np.sqrt(((data - x) ** 2).sum(axis=1)))
    return distances.idxmin(axis=1) # Which cluster a player belongs to based on the closest distance of the player's features to a cluster
```

```python
labels = get_labels(data, centroids)
```

```python
labels.value_counts()
```

```python
def new_centroids(data, labels, k):
    return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA # Turns 5d data to 2d data
import matplotlib.pyplot as plt
from IPython.display import clear_output 
```

```python
def plot_clusters(data, labels, centroids, iteration):

    # Transform data to 2d data
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels)
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1])
    plt.show()
```

```python
max_iterations = 100
k = 3

centroids = random_centroids(data, k)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids): # Loop until max iterations or until centroids don't change position
    old_centroids = centroids

    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, k)
    plot_clusters(data, labels, centroids, iteration)
    iteration += 1
```
